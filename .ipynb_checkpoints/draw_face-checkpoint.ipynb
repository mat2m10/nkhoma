```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.path import Path
from matplotlib.patches import PathPatch


# ----------------------------
# Core geometry: rounded skewed "face" patch
# ----------------------------
def draw_skewed_rectangle(width, height, angle_top_to_bottom,
                          rounding_top, rounding_bottom,
                          cx=0.5, cy=0.5,
                          facecolor="none", edgecolor="black", lw=3,
                          zorder=10):
    """
    Rounded skewed rectangle (trapezoid-ish) built as a PathPatch.

    width: top width (data units, in your 0..1 axes)
    height: height (data units)
    angle_top_to_bottom: degrees, side tilt from vertical; negative => bottom narrower
    rounding_top: radius for TL/TR corners (data units)
    rounding_bottom: radius for BR/BL corners (data units)

    Returns: PathPatch
    """
    theta = np.deg2rad(angle_top_to_bottom)
    dx = height * np.tan(theta)

    w_top = width
    w_bot = width + 2 * dx

    y_top = cy + height / 2
    y_bot = cy - height / 2

    tl = np.array([cx - w_top / 2, y_top])
    tr = np.array([cx + w_top / 2, y_top])
    br = np.array([cx + w_bot / 2, y_bot])
    bl = np.array([cx - w_bot / 2, y_bot])

    corners = [tl, tr, br, bl]  # clockwise

    # Per-corner rounding: TL,TR,BR,BL
    r_corner = np.array([rounding_top, rounding_top, rounding_bottom, rounding_bottom], dtype=float)

    # Clamp each corner radius based on adjacent edge lengths
    edge_len = [np.linalg.norm(corners[(i + 1) % 4] - corners[i]) for i in range(4)]
    for i in range(4):
        adj_min = min(edge_len[i - 1], edge_len[i])  # edges meeting at corner i
        r_corner[i] = max(0.0, min(float(r_corner[i]), 0.49 * float(adj_min)))

    p_in, p_out = [], []
    for i in range(4):
        p = corners[i]
        p_prev = corners[i - 1]
        p_next = corners[(i + 1) % 4]
        r = float(r_corner[i])

        u_prev = (p_prev - p)
        u_prev = u_prev / np.linalg.norm(u_prev)

        u_next = (p_next - p)
        u_next = u_next / np.linalg.norm(u_next)

        # incoming point is towards previous corner; outgoing is towards next corner
        p_in.append(p + u_prev * r)
        p_out.append(p + u_next * r)

    verts, codes = [], []

    # Start just after TL, going toward TR
    verts.append(tuple(p_out[0]))
    codes.append(Path.MOVETO)

    # Walk TR, BR, BL: line to incoming point, then quadratic corner to outgoing point
    for i in [1, 2, 3]:
        verts.append(tuple(p_in[i]));     codes.append(Path.LINETO)
        verts.append(tuple(corners[i]));  codes.append(Path.CURVE3)  # control
        verts.append(tuple(p_out[i]));    codes.append(Path.CURVE3)  # end

    # Close around TL
    verts.append(tuple(p_in[0]));     codes.append(Path.LINETO)
    verts.append(tuple(corners[0]));  codes.append(Path.CURVE3)
    verts.append(tuple(p_out[0]));    codes.append(Path.CURVE3)

    verts.append((0.0, 0.0)); codes.append(Path.CLOSEPOLY)

    path = Path(verts, codes)
    return PathPatch(
        path,
        facecolor=facecolor,
        edgecolor=edgecolor,
        linewidth=lw,
        joinstyle="round",
        capstyle="round",
        zorder=zorder
    )


# ----------------------------
# Separator line: arc trimmed by face boundary
# ----------------------------
def _sample_ellipse_arc(cx, cy, rx, ry, t1_deg, t2_deg, n=2500):
    th = np.deg2rad(np.linspace(t1_deg, t2_deg, n))
    return np.c_[cx + rx * np.cos(th), cy + ry * np.sin(th)]


def add_separator_line(ax, face_patch,
                       center=(0.5, 0.5),
                       vertical_pos_px=80,
                       size_px=600,
                       curvature=0.08,      # in data units (0..1). bigger = more arched
                       n=400,
                       lw=10,
                       color="black",
                       zorder=25,
                       draw=True):
    """
    Separator line that ALWAYS intersects the face sides.

    It:
      - computes y_sep = cy + vertical_pos_px/size_px
      - finds left/right x where the face boundary crosses y_sep
      - draws a quadratic Bezier between them with a controllable arch (curvature)

    curvature:
      - positive -> arches upward
      - negative -> arches downward
    """
    cx, cy = center
    y_sep = cy + float(vertical_pos_px) / float(size_px)

    # Face path in data coords
    face_path = face_patch.get_path().transformed(face_patch.get_transform())
    verts = face_path.vertices

    # Collect intersections of polygon segments with horizontal line y=y_sep
    xs = []
    for (x1, y1), (x2, y2) in zip(verts[:-1], verts[1:]):
        # Skip nearly horizontal segments (avoid noise)
        if abs(y2 - y1) < 1e-12:
            continue
        # Check if segment crosses y_sep
        if (y1 <= y_sep <= y2) or (y2 <= y_sep <= y1):
            t = (y_sep - y1) / (y2 - y1)
            if 0.0 <= t <= 1.0:
                xs.append(x1 + t * (x2 - x1))

    if len(xs) < 2:
        raise ValueError(
            "Separator y does not intersect the face. "
            "Try a different vertical_pos_px (move it up/down)."
        )

    x_left, x_right = float(min(xs)), float(max(xs))

    # Quadratic Bezier control point at mid x, arched by curvature
    x_mid = 0.5 * (x_left + x_right)
    ctrl = np.array([x_mid, y_sep + curvature])

    # Sample Bezier
    t = np.linspace(0, 1, n)
    p0 = np.array([x_left, y_sep])
    p2 = np.array([x_right, y_sep])

    pts = (1 - t)[:, None] ** 2 * p0 + 2 * (1 - t)[:, None] * t[:, None] * ctrl + t[:, None] ** 2 * p2

    if draw:
        ax.add_line(mlines.Line2D(
            pts[:, 0], pts[:, 1],
            lw=lw, color=color,
            solid_capstyle="round",
            zorder=zorder
        ))

    return pts



# ----------------------------
# Convenience: setup axes (helps for consistent sizing)
# ----------------------------
def make_logo_axes(fig_size=(6, 6), xlim=(0, 1), ylim=(0, 1)):
    fig, ax = plt.subplots(figsize=fig_size)
    ax.set_xlim(*xlim)
    ax.set_ylim(*ylim)
    ax.set_aspect("equal")
    ax.axis("off")
    return fig, ax


# ----------------------------
# Example usage (your current test)
# ----------------------------
if __name__ == "__main__":
    fig, ax = make_logo_axes(fig_size=(6, 6))

    # (Note: you wrote /3; classic golden ratio is /2. Use whichever you intend.)
    phi = (1 + 5**0.5) / 3

    height = 0.8
    width = height * phi
    line_width_biggest = 10

    face = draw_skewed_rectangle(
        width=width, height=height,
        angle_top_to_bottom=-12,
        rounding_top=0.12,      # <- start smaller than 0.5 (data units)
        rounding_bottom=0.18,   # <- and tune
        cx=0.5, cy=0.5,
        facecolor="white",
        edgecolor="black",
        lw=line_width_biggest,
        zorder=10
    )
    ax.add_patch(face)

    add_separator_line(
        ax,
        face_patch=face,
        center=(0.5, 0.5),
        vertical_pos_px=90,
        size_px=600,
        rx=0.55,
        ry=0.18,
        arc_deg=(200, 340),
        lw=line_width_biggest,
        color="black",
        zorder=25
    )

    # Give yourself padding so thick strokes don't clip
    ax.set_xlim(-0.1, 1.1)
    ax.set_ylim(-0.1, 1.1)

    plt.show()
```
